## Logistic Regression

```python
predictor_columns=['age', 'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate',
       'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed',
       'admin.', 'blue-collar', 'entrepreneur', 'housemaid', 'management',
       'retired', 'self-employed', 'services', 'student', 'technician',
       'unemployed', 'unknown', 'divorced', 'married', 'single', 'unknown',
       'basic.4y', 'basic.6y', 'basic.9y', 'high.school', 'illiterate',
       'professional.course', 'university.degree', 'unknown', 'no', 'unknown',
       'yes', 'no', 'unknown', 'yes', 'no', 'unknown', 'yes', 'cellular',
       'telephone', 'apr', 'aug', 'dec', 'jul', 'jun', 'mar', 'may', 'nov',
       'oct', 'sep', 'fri', 'mon', 'thu', 'tue', 'wed', 'failure',
       'nonexistent', 'success']
target_columns=['y']
```

```python
x=bank[predictor_columns]
y=bank[target_columns]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20
                                                    , random_state=101)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## Building the model

```python
from sklearn.linear_model import LogisticRegression

```

```python
logmodel= LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
predictions=logmodel.predict(x_test)
```

```python
from sklearn.feature_selection import RFE
rfe=RFE(logmodel,15)
rfe=rfe.fit(x_train,y_train)
```

```python

list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python

col=x_train.columns[rfe.support_]
```

```python
#select only the columns selected by RFE
x_train=x_train[col]
```

```python
from sklearn import metrics
```

```python

fpr,tpr,thresholds=metrics.roc_curve(y_test, positive_score, pos_label=1)
```

```python
fpr,tpr,thresholds
```

```python
def plot_roc_curve(fpr,tpr,label=None ):
    plt.plot(fpr,tpr,label=label)
    plt.plot([1,0],[1,0],'k--')
    plt.axis([0,1,0,1])
    plt.title('ROC curve')
    plt.xlabel(' false positive rate')
    plt.ylabel(' True positive rate')
    plt.legend()
    
plot_roc_curve(fpr,tpr,"logistic_regression")
plt.show()
```
